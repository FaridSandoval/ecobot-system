### Hacer fine tunning de Yolo V8

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

# Train the model
# Se debe de definir la ruta de el archivo .zip que se descarga en RoboFlow, ademas de el numero de epocas y el tamanio de las imagenes que por defecto es 640
results = model.train(data="/content/drive/MyDrive/uao/dataset/xray_chest/xr chest.v2i.yolov8.zip", epochs=100, imgsz=640)
```
